```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random

from google.colab import drive
drive.mount('/content/drive')

# Device Information DF
df=pd.read_excel('drive/MyDrive/added_features_testing.xlsx')
df.head()
```

```python
# Create Device List
adid_df = df.dropna(subset=["Device ID"])
adid_column = adid_df[['Device ID']].drop_duplicates()
adid_list = adid_column['Device ID'].tolist()
random.shuffle(adid_list)
print(len(adid_list))
```

```python
# ADID Location DF
df1=pd.read_csv('drive/MyDrive/9638_Plainfield.csv')
```

```python
# Create Columns for Dates and Time
testing = df1.copy()
testing['Signal Date']= pd.to_datetime(testing['Signal Date'])

testing['Dates'] = pd.to_datetime(testing['Signal Date']).dt.date
testing['Time'] = pd.to_datetime(testing['Signal Date']).dt.time

testing.head()
```

```python
# Aggregate 24 Hour Periods
count_minute_df = testing.groupby(['ADID','Dates'])['Signal Date'].agg(Total_Session_Times=np.ptp,Session_Counts='count').reset_index()
count_minute_df['Total_Session_Minutes']= count_minute_df.Total_Session_Times.dt.total_seconds()/60
count_minute_df = count_minute_df.drop(columns = ['Total_Session_Times'])

count_minute_df.head()
```

```python
# Aggregate Total MMinutes and Counts Per ADID
f = {'Total_Session_Minutes':['sum'], 'Session_Counts':['sum']}
count_minute_df = count_minute_df.groupby('ADID').agg(f).reset_index()
count_minute_df.columns = count_minute_df.columns.droplevel(1)
count_minute_df.head()
```

```python
# Join Minutes and Cunts with Location DF
df1 = pd.merge(df1,count_minute_df,how='inner',on='ADID')
print(len(df1))
df1.head()
```

```python
# Create DF and List of ADID's
adid_df1 = df1[['ADID']]
print(len(adid_df1))
next = adid_df1.drop_duplicates()
adid_list_df1 = next['ADID'].tolist()
print(len(adid_list_df1))
```

```python
# Randomly Attach Device to ADID
new_list = []
for x in adid_list_df1:
  random_val = random.choice(adid_list)
  new_list.append(random_val)
print(len(new_list))
```

```python
# Add Device ID's to ADID DF
next['Device ID'] = new_list
next.head()
```

```python
# Join ADID DF with Device DF
curious = pd.merge(df1,next,how='inner',on='ADID')
print(len(curious))
curious.head()
```

```python
# Join ADID Location DF with Device DF
next_curious = pd.merge(curious,adid_df,how='inner',on='Device ID')
print(len(next_curious))
next_curious.head()
```
