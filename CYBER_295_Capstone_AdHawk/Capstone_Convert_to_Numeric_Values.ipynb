```python
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
import datetime
import dateutil.parser
import ipaddress
import statistics

from datetime import datetime
from sklearn import metrics
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.mixture import GaussianMixture
from matplotlib.colors import LogNorm
```

```python
df=pd.read_csv('drive/MyDrive/first_add_features_df.csv')
df = df.drop_duplicates()
df.head()
```

```python
# Convert Category Columns to Code
df['Browser Family'] = df['Browser Family'].astype('category')
df['Browser_Family_Codes'] = df['Browser Family'].cat.codes

df['OS Family'] = df['OS Family'].astype('category')
df['OS_Family_Codes'] = df['OS Family'].cat.codes

df['Device Family'] = df['Device Family'].astype('category')
df['Device_Family_Codes'] = df['Device Family'].cat.codes

df['Device Brand'] = df['Device Brand'].astype('category')
df['Device_Brand_Codes'] = df['Device Brand'].cat.codes

df['Device Model'] = df['Device Model'].astype('category')
df['Device_Model_Codes'] = df['Device Model'].cat.codes

df['EU'] = df['EU'].astype('category')
df['EU_Codes'] = df['EU'].cat.codes

df['US'] = df['US'].astype('category')
df['US_Codes'] = df['US'].cat.codes

df['Country Code'] = df['Country Code'].astype('category')
df['Country_Codes'] = df['Country Code'].cat.codes

df['Gender'] = df['Gender'].astype('category')
df['Gender_Codes'] = df['Gender'].cat.codes

df['Marital Status'] = df['Marital Status'].astype('category')
df['Marital_Status_Codes'] = df['Marital Status'].cat.codes

df['Education Level'] = df['Education Level'].astype('category')
df['Education_Level_Codes'] = df['Education Level'].cat.codes

df['Generational'] = df['Generational'].astype('category')
df['Generational_Codes'] = df['Generational'].cat.codes

df['Parent'] = df['Parent'].astype('category')
df['Parent_Codes'] = df['Parent'].cat.codes

df['Country Code2'] = df['Marital Status'].astype('category')
df['Country_Code2_Codes'] = df['Marital Status'].cat.codes

df['Category'] = df['Category'].astype('category')
df['Category_Codes'] = df['Category'].cat.codes

df['Discription'] = df['Discription'].astype('category')
df['Discription_Codes'] = df['Discription'].cat.codes

# View the converted DataFrame
print(df[['Browser_Family_Codes', 'OS_Family_Codes', 'Device_Family_Codes', 'Device_Brand_Codes',
          'Device_Model_Codes', 'EU_Codes', 'US_Codes', 'Country_Codes',
          'Gender_Codes', 'Marital_Status_Codes', 'Education_Level_Codes', 'Generational_Codes',
          'Parent_Codes', 'Country_Code2_Codes', 'Category_Codes', 'Discription_Codes']].head())
```

```python
# Convert Dates and Time to Numeric Values
df['signal_date_int'] = df['Signal Date'].apply(lambda x: int(datetime.strptime(x, '%m/%d/%Y %H:%M').timestamp()))
df['created_on_int'] = df['Created on'].apply(lambda x: int(dateutil.parser.isoparse(x).timestamp()))
df['last_update_int'] = df['Last Update'].apply(lambda x: int(dateutil.parser.isoparse(x).timestamp()))
df['first_signal_int'] = df['First Signal'].apply(lambda x: int(dateutil.parser.isoparse(x).timestamp()))
df['last_signal_int'] = df['Last Signal'].apply(lambda x: int(dateutil.parser.isoparse(x).timestamp()))

df[['Signal Date', 'signal_date_int', 'Created on', 'created_on_int', 'Last Update', 'last_update_int',
    'First Signal', 'first_signal_int', 'Last Signal', 'last_signal_int']].head()
```

```python
# Convert Miscelaneuos Values to Numeric
df['annual_income_int'] = df['Annual Income'].apply(lambda x: x if isinstance(x, float) else statistics.median(map(int, x.split("-"))))
df['os_version_int'] = df['OS Version'].apply(lambda x: x if isinstance(x, float) else int(x.split('.', 1)[0]))
df[[ 'Annual Income', 'annual_income_int', 'OS Version', 'os_version_int']].head()
```

```python
# Drop Unused Features
df = df.drop(columns = ["Latitude", "Longitude", "Address", "ADID", "Device ID", "Full","Column3",
                        'Browser Family', 'OS Family', 'Device Family',
                        'Device Brand', 'Device Model', 'EU', 'US',
                        'Country Code', 'Gender', 'Marital Status',
                        'Education Level', 'Generational', 'Parent',
                        'Country Code2', 'Category', 'Discription',
                        'Timestamp', 'Last seen', 'Created on', 'Last Update', 'First Signal',
                        'IP Address', 'Annual Income', 'Browser Version', 'Signal Date'
                        ,'Last Signal', 'OS Version'])
```
